```python
def time_to_regional(TA):
    RA = {
        "A": {'0, x=0, y=0', '0, x=1, y=0', '1, x<1, y=0', '1, x<1, y<1', '1, x<1, y=1', '1, x<1, y>1', '2, x<1, y=2', '2, x=1, y=2', '2, x>1, y=2'},
        "X": TA["X"].copy(),
        "A0": {'0, x=0, y=0', '0, x=1, y=0'},
        "E": {
            '0, x=0, y=0': {
                'a': {'0, x=1, y=0'},
                'b': {'1, x<1, y=0', '1, x<1, y<1', '1, x<1, y=1', '1, x<1, y>1'}
            },
            '1, x<1, y=0': {
                'a': {'0, x=0, y=0', '0, x=1, y=0'},
                'b': {'2, x<1, y=2', '2, x=1, y=2', '2, x>1, y=2'}
            },
            '1, x<1, y<1': {
                'b': {'2, x<1, y=2', '2, x=1, y=2', '2, x>1, y=2'}
            },
            '1, x<1, y=1': {
                'b': {'2, x<1, y=2', '2, x=1, y=2', '2, x>1, y=2'}
            },
            '1, x<1, y>1': {
                'b': {'2, x<1, y=2', '2, x=1, y=2', '2, x>1, y=2'}
            }
        }
    }

    return RA


TA = {
    "A": {'0', '1', '2'},
    "X": {'a', 'b'},
    "A0": "0",
    "Y": {'x', 'y'},
    "E": {
        "0": {
            "a, x=1": "0",
            "b, x<1": "1"
        },
        "1": {
            "a, x<2": "0",
            "b, y=2": "2"
        },
        "2": {
            "a, x=1, y=0": "2",
            "b, x=1, y=0": "2"
        }
    },
    "F": "1"
}

RA = time_to_regional(TA)
print(RA)
```
