```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sales = pd.read_csv('https://raw.githubusercontent.com/sbohatko/MastersLabs/main/applied%E2%80%93application-of-neural-networks/lab2/video_games_sales.csv')
sales.head()
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/sbohatko/MastersLabs/main/applied%E2%80%93application-of-neural-networks/lab2/iris.csv')
iris.head()
```

# Linear/Graph/Pie charts

```python
# Drop rows with missing values in 'Genre' or 'EU_Sales' columns
cleanedSales = sales.dropna(subset=['Genre', 'EU_Sales'])

# Group by 'Genre' and sum the 'EU_Sales' for each genre
genre_sales = cleanedSales.groupby('Genre')['EU_Sales'].sum().reset_index()

# Plotting a simple line graph
plt.figure(figsize=(12, 6))
plt.plot(genre_sales['Genre'], genre_sales['EU_Sales'], marker='o', linestyle='-')
plt.title('EU Sales vs Genre')
plt.xlabel('Genre')
plt.ylabel('EU Sales (in millions)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.grid(True)
plt.show()


# Plot a bar chart for 'EU_Sales' by 'Genre'
plt.figure(figsize=(12, 6))
plt.bar(genre_sales['Genre'], genre_sales['EU_Sales'], color='skyblue')
plt.xlabel('Genre')
plt.ylabel('EU Sales (in millions)')
plt.title('EU Sales by Genre')
plt.xticks(rotation=45, ha='right')
plt.show()

# Plot a pie chart for 'EU_Sales' by 'Genre'
plt.figure(figsize=(10, 6))
plt.pie(genre_sales['EU_Sales'], labels=genre_sales['Genre'], autopct='%1.1f%%', startangle=140)
plt.title('EU Sales Distribution by Genre')
plt.show()
```

# Seaborn Scatter/Line/Bar plots

```python

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set1', s=80)
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()

# Line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=iris, x='sepal_length', y='petal_length', hue='species', style='species', markers=True, dashes=False)
plt.title('Line Plot of Sepal Length vs Petal Length')
plt.show()

# Bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=iris, x='species', y='petal_width', errorbar=None, palette='viridis')
plt.title('Bar Plot of Petal Width by Species')
plt.show()
```

```python

# Create a figure with three subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

# Scatter plot on the first subplot
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='Set1', s=80, ax=axes[0])
axes[0].set_title('Scatter Plot of Sepal Length vs Sepal Width')

# Line plot on the second subplot
sns.lineplot(data=iris, x='sepal_length', y='petal_length', hue='species', style='species', markers=True, dashes=False, ax=axes[1])
axes[1].set_title('Line Plot of Sepal Length vs Petal Length')

# Bar plot on the third subplot
sns.barplot(data=iris, x='species', y='petal_width', errorbar=None, palette='viridis', ax=axes[2])
axes[2].set_title('Bar Plot of Petal Width by Species')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
